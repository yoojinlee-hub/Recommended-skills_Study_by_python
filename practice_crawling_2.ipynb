```python
import requests
from bs4 import BeautifulSoup

# 코드를 작성하세요
rating_pages = []
pages = range(1, 50)
for page in pages:
    url = "https://movie.naver.com/movie/point/af/list.naver?&page={}".format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'html.parser')
    rating_pages.append(soup.find_all("td", {"class":"title"})) #리스트에 요소 추가(append)
# 테스트 코드
print(len(rating_pages)) # 가져온 총 페이지 수 
print(rating_pages[0]) # 첫 번째 페이지의 HTML 코드
```
