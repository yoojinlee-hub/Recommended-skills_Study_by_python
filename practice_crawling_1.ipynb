```python
import requests

# 코드를 작성하세요
rating_pages = []
years = range(2010, 2013)
months = range(1, 13)
weeks = range(0, 5)
for year in years:
    for month in months:
        for week in weeks:
            url = "https://workey.codeit.kr/ratings/index?year={}&month={}&weekIndex={}".format(year, month, week)
            response = requests.get(url)
            rating_pages.append(response.text) #리스트에 요소 추가(append)
# 테스트 코드
print(len(rating_pages)) # 가져온 총 페이지 수 
print(rating_pages[0]) # 첫 번째 페이지의 HTML 코드
```

```python
import requests
from bs4 import BeautifulSoup

# HTML 코드 받아오기
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
response = requests.get(url).text.encode('utf-8')
response = BeautifulSoup(response, 'html.parser')

print(response)
```

```python
result01 = list(response.find('div',class_='tit3').stripped_strings)
print(result01)
```

```python
result02 = list(response.find_all('div',class_='tit3'))
print(result02)
```
